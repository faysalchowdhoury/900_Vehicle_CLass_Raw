```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'your_file.xlsx' with the actual file name and 'SheetName' with the sheet you want to read
file_path = 'new_graph.xlsx'
sheet_name = 'Sheet3'

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(
    x="Vehicle Class Number",
    y="Number of Vehicles",
    data=df,
    orient="v",  # Orient the bar plot vertically
)

# Add title, labels, and rotate x-axis labels for better visibility
plt.title("Number of Vehicles per Vehicle Class")
plt.xlabel("Vehicle Class Number")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=90)
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'your_file.xlsx' with the actual file name and 'SheetName' with the sheet you want to read
file_path = 'new_graph.xlsx'
sheet_name = 'Sheet3'

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Create a count plot using Seaborn
plt.figure(figsize=(12, 6))
sns.countplot(x="Rows", data=df)
plt.title("Number of Vehicles per Row")
plt.xlabel("Row")
plt.ylabel("Count")
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
import pandas as pd

# Replace 'your_file.xlsx' with the actual file name
file_path = 'new_graph1.xlsx'

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(file_path)

# Melt the DataFrame to move rows to a new column
df_melted = df.melt(id_vars=["Vehicle Class Number"], value_vars=df.columns[1:], var_name="Rows", value_name="Number of Vehicles")

# Print the first 5 rows of the melted DataFrame
print(df_melted.head())
```

```python
# Save the melted DataFrame to an Excel file
output_file_path = 'melted_data.xlsx'
df_melted.to_excel(output_file_path, index=False)

print("Melted DataFrame saved to Excel file:", output_file_path)
```
